```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)     # para poder visualizar todas las columnas de los DF
```

```python
# lectura de archivo y lo convierte en un DF
df_Flight = pd.read_csv(r"C:\Users\Usuario\Desktop\Ejercicios Modulo3\files/Customer Flight Activity.csv")
df_Loyalty = pd.read_csv(r"C:\Users\Usuario\Desktop\Ejercicios Modulo3\files/Customer Loyalty History.csv")
```

```python
# unión mediante "Loyalty number" porque ambos contienen info de un mismo cliente(ID)
df_merge = df_Flight.merge(df_Loyalty, how = 'left', left_on = 'Loyalty Number', right_on = 'Loyalty Number')
df_merge.head()

df_final = df_merge.copy()
```

```python
# reemplazo los nulos por 0

columnas_imputadas = ["Salary", "Cancellation Year", "Cancellation Month"]

df_final[columnas_imputadas] = df_final[columnas_imputadas].fillna(0) # teemplaza NaN por 0
```

```python
# duplicados: hay 3936 filas donde el mismo cliente aparece en el mismo año y mes

df_final.duplicated(
    subset=["Loyalty Number", "Year", "Month"]
).sum()
```

```python
# eliminación de duplicados: mantengo la primera fila de cada cliente/año/mes
df_final = df_final.drop_duplicates(subset=["Loyalty Number", "Year", "Month"], keep='first')
```

```python
# elimino la columna 'Province'
df_final.drop(columns=['Province'], errors='ignore', inplace = True) #si no existe, no hace nada 
df_final.columns
```

```python
# uno los valores de las columnas Year y Month para añadirlos a una nueva columna: Flights Date

df_final['Flights Date'] = pd.to_datetime(df_final[['Year', 'Month']].assign(Day=1))

# elimino las columnas Year y Month

df_final = df_final.drop(columns=['Year', 'Month'])
```

```python
# idem
df_final['Enrollment Date'] = pd.to_datetime(df_final[['Enrollment Year', 'Enrollment Month']].rename(columns={
'Enrollment Year': 'Year','Enrollment Month': 'Month'}).assign(Day=1))

df_final = df_final.drop(columns=['Enrollment Year', 'Enrollment Month'])
```

```python
# idem: para poder poner un formato fecha debo reemplazar los 0 por NaN

df_final[['Cancellation Year', 'Cancellation Month']] = df_final[['Cancellation Year', 'Cancellation Month']].replace(0, pd.NA)

df_final['Cancellation Date'] = pd.to_datetime(             # fecha, errors = 'coerce' convierte las filas con NaN en NaT
    df_final.assign(
        year=df_final['Cancellation Year'],
        month=df_final['Cancellation Month'],
        day=1
    )[["year","month","day"]],
    errors="coerce"
)

df_final = df_final.drop(columns=['Cancellation Year', 'Cancellation Month'])
```

```python
# VERIFICAR la consistencia de los datos
# métricas

df_final = df_final[df_final['Flights Booked'] >= 0]
df_final = df_final[df_final['Total Flights'] >= 0]

df_final = df_final[df_final['Distance'] >= 0]


df_final = df_final[df_final['Points Accumulated'] >= 0]
df_final = df_final[df_final['Points Redeemed'] >= 0]


df_final = df_final[df_final['Salary'] >= 0]

# fecha correcta

df_final = df_final[df_final['Flights Date'] >=  '2000-01-01']
df_final = df_final[df_final['Enrollment Date'] >=  '2000-01-01']
df_final = df_final[df_final['Cancellation Date'] >=  '2000-01-01']
```

```python
# tipo de datos: cambio Points Accumulated porque tiene sentido que sea int igual que Point Redeemed
df_final.info()
```

```python
# cambio Points Accumulated porque tiene sentido que sea int (igual que Point Redeemed)
df_final['Points Accumulated'] = df_final['Points Accumulated'].astype('int64')
```

```python
# FASE 2
# Análisis de variables numéricas:

variables_numericas = ['Flights Booked', 'Flights with Companions', 'Total Flights','Distance', 'Points Accumulated', 'Points Redeemed', 
'Dollar Cost Points Redeemed', 'Salary', 'CLV']

resultado = df_final[variables_numericas].agg(["mean", "median", "std", "var", "min", "max"])
print(resultado)
```

```python
df_final.describe().T
```

```python
# valores atípicos (outliers) por variable

for i in variables_numericas:

    Q1 = df_final[i].quantile(0.25)
    Q3 = df_final[i].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df_final[(df_final[i] < Q1 - 1.5*IQR) | (df_final[i] > Q3 + 1.5*IQR)]
    print(f"{i}: {len(outliers)} outliers")
```

```python
# correlación entre variables numéricas: que tan relacionadas están 2 variables numéricas 
# Ejemplo: Distance + Points accumulated --> cuanto más viaje más puntos acumula

correlacion = df_final[variables_numericas].corr()
print(correlacion)
```

```python
# Análisis de variables categóricas:

variables_categoricas = ["Gender", "Education", "Marital Status", "Loyalty Card", "Enrollment Type"]

for i in variables_categoricas:
    print(i)
    
    # nº clientes por categoría 
    print(df_final[i].value_counts())    
    
    # % clientes por categoría
    print(df_final[i].value_counts(normalize=True).mul(100).round(2))  
```

```python
# FASE 3. VISUALIZACIÓN

# 3.1. ¿Cómo se distribuye la cantidad de vuelos reservados por mes durante el año?
df_final["Month"] = df_final["Flights Date"].dt.month
df_final["Year"] = df_final["Flights Date"].dt.year

vuelos_por_mes = df_final.groupby("Month")["Total Flights"].sum()
vuelos_por_mes.plot(kind="bar", color = "blue")


plt.xlabel("Vuelos reservados")
plt.ylabel("Mes")
plt.title("Nº de vuelos por mes")
plt.show()
```

```python
# 3.2 . ¿Existe una relación entre la distancia de los vuelos y los puntos acumulados por los cliente?

sns.scatterplot(
    x = "Distance",
    y = "Points Accumulated",
    data =df_final
)

plt.xlabel("Distancia")
plt.ylabel("Puntos acumulados")
plt.title("Distancia vs Puntos acumulados")
plt.show()

# interpretación: cada punto en el gráfico representa un vuelo. Un punto en 5000 seria un vuelo de 1500 km que dio 425 puntos
# veo una tendencia ascendente (a mayor distancia + puntos acumulados = relación positiva)
```

```python
# 3.4. ¿Cómo se compara el salario promedio entre los diferentes niveles educativos de los clientes?
# Promedio

# Filtramos niveles educativos con salario > 0
df_salario = df_final[df_final["Salary"] > 0]

salario_educacion_avg = df_salario.groupby("Education")["Salary"].mean()
salario_educacion_avg
```

```python
salario_educacion_avg.plot(kind="bar", color="teal", edgecolor="black")

plt.xlabel("Nivel educativo")
plt.ylabel("Promedio salario")
plt.title("Salario promedio según nivel educativo")
plt.show()
```

```python
# 3.5 ¿Cuál es la proporción de clientes con diferentes tipos de tarjetas de fidelidad?
# Conteo de clientes por tipo de tarjeta
tarjetas_count = df_final["Loyalty Card"].value_counts()

# Proporción (porcentaje)
tarjetas_prop = df_final["Loyalty Card"].value_counts(normalize=True) * 100

tarjetas_prop
```

```python
# 3.6. ¿Cómo se distribuyen los clientes según su estado civil y género?
plt.figure(figsize=(10,6))

sns.countplot(
    x="Marital Status",
    hue="Gender",
    data=df_final
)

plt.xlabel("Estado civil")
plt.ylabel("Número de clientes")
plt.title("Distribución de clientes según estado civil y género")
plt.xticks(rotation=45)
plt.show()
```

```python
# FASE 4

# Filtrar columnas relevantes
df_vuelos = df_final[["Flights Booked", "Education"]]

# Eliminar valores nulos o negativos
df_vuelos = df_vuelos[df_vuelos["Flights Booked"] >= 0]
df_vuelos = df_vuelos.dropna(subset=["Flights Booked", "Education"])
```

```python
# Agrupar por Education y calcular estadísticas descriptivas
estadisticas_vuelos = df_vuelos.groupby("Education")["Flights Booked"].agg(
    promedio="mean",
    desviacion_estandar="std",
    minimo="min",
    maximo="max",
    cuenta="count"
)

estadisticas_vuelos
```

```python


estadisticas_vuelos["promedio"].plot(kind="bar", color="skyblue")
plt.xlabel("Nivel educativo")
plt.ylabel("Promedio de vuelos reservados")
plt.title("Número promedio de vuelos reservados según nivel educativo")
plt.show()
```
